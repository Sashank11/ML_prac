### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # width and height

#3. Plot Data
ax.plot(x,y)

#4. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

#5. Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy Arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data
x = np.linspace(0, 10, 100) # Does even space intervals
x[:10]
```

```python
# Plot the Data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x) );
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter" : 10,
                     "Peanut Butter" : 8,
                     "Cashew butter" : 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values()) 

#python bit: postion argument shouldnt be followed with keyword argument
#usualy either both or the 2nd argument should have a keyword argument
ax.set(title = "Dan's Nut Butter Store",
      ylabel = "Price ($)"); # add a semicolon to the last line where ax is called to remove that nasty line.
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
# Make some data for histogram
x = np.random.randn(1000) # normal distribution bell-curve like
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows =2, ncols =2, figsize = (8,8))

# Plot to each different axes
ax1.plot(x, x/2)
ax2.scatter(x, np.exp(x))
ax3.hist(np.random.randn(1000))
ax4.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2, ncols =2, figsize =(8,8))
# plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(x, np.exp(x))
ax[1,0].hist(np.random.randn(1000))
ax[1,1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

### Pyplot vs Matplotlib OO method
* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advanced, use OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind = "scatter", x = "age", y = "chol", c = "target");
```

```python
# OO Method mixed with pyplot method
fix, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
                x = "age",
                y = "chol",
                c = "target",
                ax = ax);
#ax.set_xlim([45,100]);
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize = (8,5))

# Plot the Data
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"]);

#Customize
ax.set(title = "Heart Diesease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

#Add a horizontal Line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--" );
```

```python
heart_disease.head()
```

```python
#OO method from scratch
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10), sharex = True)

# Plot the Data
scatter1 = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"]);
scatter2 = ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"]);
#Customize
ax0.set(title = "Heart Diesease and Cholesterol Levels",
       ylabel = "Cholesterol");
ax1.set(title = "Heart Diesease and Max Heart Rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate");

# Add a legend
ax0.legend(*scatter1.legend_elements(), title = "Target");
ax1.legend(*scatter2.legend_elements(), title = "Target");

#Add a horizontal Line
ax0.axhline(over_50["chol"].mean(),
           linestyle = "--" );
ax1.axhline(over_50["thalach"].mean(),
           linestyle = "--" );

#Add title to the figure
fig.suptitle("Heart Disease Analysiss", fontsize = 16, fontweight = "bold");
```
